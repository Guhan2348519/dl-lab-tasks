```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/drive/MyDrive/Tweets.csv')
df.head(5)
```

```python
actualSentiment = df['airline_sentiment']
positiveNegative = actualSentiment.map(lambda x: 1 if x=='positive' else 0)
df['Sentiment'] = positiveNegative
df['Sentiment'].value_counts()
```

```python
# Preprocessing
# Assuming 'text' column contains the tweets and 'airline_sentiment' column contains sentiment labels
texts = df['text'].astype(str)
labels = df['airline_sentiment']
maxlen = 100  # Max length of each sequence
max_words = 10000  # Vocabulary size

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))

data = pad_sequences(sequences, maxlen=maxlen)

labels = pd.get_dummies(labels)

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
# RNN Model
model_rnn = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(max_words, 32),
    tf.keras.layers.SimpleRNN(32),
    tf.keras.layers.Dense(3, activation='softmax')  # 3 output classes: negative, neutral, positive
])

model_rnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(model_rnn.summary())

history_rnn = model_rnn.fit(X_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2)  # 20% of training data for validation
```

This code creates an RNN model with an embedding layer, a SimpleRNN layer, and a dense output layer with a softmax activation function. It compiles the model using the Adam optimizer and categorical cross-entropy loss function, then trains it for 10 epochs on the training data with a validation split of 20%.

```python
# LSTM Model
model_lstm = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(max_words, 32),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(3, activation='softmax')
])

model_lstm.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print(model_lstm.summary())

history_lstm = model_lstm.fit(X_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.2)
```

You created an LSTM model with embedding and dense layers, compiled it with Adam optimizer, and trained it for 10 epochs on training data with a validation split of 20%.LSTM (Long Short-Term Memory) is a type of recurrent neural network (RNN) architecture commonly used for processing and making predictions based on sequential data. It is designed to capture long-term dependencies and overcome the limitations of traditional RNNs in handling information over extended time lags.


```python
# Evaluate the models
loss_rnn, accuracy_rnn = model_rnn.evaluate(X_test, y_test)
print("RNN Test Accuracy:", accuracy_rnn)

loss_lstm, accuracy_lstm = model_lstm.evaluate(X_test, y_test)
print("LSTM Test Accuracy:", accuracy_lstm)
```

The RNN model attained a test accuracy of approximately 71.31%, while the LSTM model achieved a slightly higher test accuracy of about 76.91%. Despite their differences in accuracy, both models demonstrated relatively similar performance on the test dataset.

```python

import random

# Function to predict sentiment using the specified model and tokenizer
def predict_sentiment(text, model, tokenizer):
    tw = tokenizer.texts_to_sequences([text])  # Remove the outer list
    tw = pad_sequences(tw, maxlen=100)  # Assuming max length is 100
    prediction = model.predict(tw)  # No need to round or convert to int yet
    return prediction

# Randomly select a text from the dataset
random_index = random.randint(0, len(df) - 1)
random_text = df['text'][random_index]

# Make predictions using the RNN and LSTM models
rnn_prediction = predict_sentiment(random_text, model_rnn, tokenizer)
lstm_prediction = predict_sentiment(random_text, model_lstm, tokenizer)

# Convert predictions to labels
rnn_label = "Positive" if rnn_prediction.argmax() == 1 else "Negative"
lstm_label = "Positive" if lstm_prediction.argmax() == 1 else "Negative"

# Print the predictions
print("Random Text:")
print(random_text)
print("\nRNN Model Predicted Sentiment:", rnn_label)
print("LSTM Model Predicted Sentiment:", lstm_label)
```
